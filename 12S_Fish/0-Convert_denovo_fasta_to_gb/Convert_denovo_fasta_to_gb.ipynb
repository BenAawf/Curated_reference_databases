```python
from reprophylo import *
```

```python
rrns = Locus(char_type='dna', 
            feature_type='rRNA', 
            name='12S_r_RNA',
            aliases=['12S r RNA', '12 S rRNA', '12S rRNA', 'rrns', 's-rRNA'])
```

```python
print rrns
```

```python
loci_list = [rrns]
```

```python
!mkdir outputs
```

```python
pj = Project(loci_list, pickle='outputs/fish_denovo.pkpj',git=False)
```

```python
!ls -alt
```

```python
denovo_sequence_filenames = ['INBO_and_EVO_denovo_12S_2.fasta']
```

```python
pj.read_denovo(denovo_sequence_filenames, 'dna', format='fasta')
```

```python
for r in pj.records:
    print r.format('genbank')
```

```python
for r in pj.records:

    pj.add_feature_to_record(r.id, 'rRNA', qualifiers={'product': "12S ribosomal RNA"})
```

```python
for r in pj.records:
    print r.format('genbank')
```

```python
features_to_modify = ['denovo0_source']
pj.add_qualifier(features_to_modify, 'organism', 'Abramis brama')
features_to_modify = ['denovo0_source']
pj.add_qualifier(features_to_modify, 'db_xref', 'taxon:38527')
```
